```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Histogram
In histograms, x axis contains a variable and y axis will be a frequency of that variable

#### 80-100: Normal
#### 100-125: Pre-diabetic
#### 125-150: Diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
bins=[80,100,125,150]

plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")


plt.hist(blood_sugar, rwidth=0.95 , bins=bins , color='#a36c96')         # rwidth is the relaive width of bars compared to bins
                                                   #bins =3 means
plt.show()
```

### Mutiple data samples in a histogram

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['#805463', '#c4104f',],label=['men','women'])
plt.legend()
plt.show()
```

### histtype=step

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],color='g',rwidth=0.95,histtype='step')
plt.show()
```

### HORIZONTAL

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, height=12,color='#ded466' , orientation='horizontal')
plt.show()
```

## PIE CHART

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
color=['#3a9c35', '#62dea8','#a4de62','#de6289', '#62bbde']

plt.pie(exp_vals,labels=exp_labels , autopct='%1.1f%%', colors=color ,radius=1.5 , explode=[0,0,0,0.1,0.2]) 

                                                         #labels is an argument to use
                                                         #autopct '%f%% before f write how many decimal place you want'
plt.axis("equal")   #to make it round                    #explode if you want one of the wedges to stand out write like array 
                                                         #and inc value at the position where you want it to stand out,                                                         
plt.show()
```
