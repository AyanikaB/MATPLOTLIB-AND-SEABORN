```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### LINE CHARTS


```python
x=[1,2,3]
y=[2,4,6]

plt.title('First Graph' , fontdict={'fontname':'Comic Sans MS' , 'fontsize':14})
plt.xlabel('X axis',fontdict={'fontsize':14})
plt.ylabel('Y axis')          # can do with all
plt.plot(x,y)
```

```python
#if we want ticks to not be decimals and just integers then-



plt.xticks([0,2,4,6,])
plt.yticks([0,2,4,6,8,10])
plt.plot(x,y , label='aya' , color='red',linewidth=2 , linestyle='--',marker='.' , markersize=10 ,
         markeredgecolor='blue')       
#to add legend we need to give name ie a label similarly other props can be defined.
plt.legend()

```

```python
# shorthand operations ; fmt='[color][marker][line]'

x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y , 'r*--')  #first word of color * as marker and -- as line style
                        #check matplotlib documentation page for rules to follow
```

```python
x1=np.arange(0,4.5,0.5)
plt.plot(x1, x1**2)
```

```python
#two together
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.plot(x,y , label='a')


x1=np.arange(0,4.5,0.5)
plt.plot(x1, x1**2 , label='b')
```

```python
#same graph diff line style to depict that it stops at a place but future prediction
x1=np.arange(0,4.5,0.5)
plt.plot(x1[:6], x1[:6]**2 , 'r')
plt.plot(x1[5:], x1[5:]**2 , 'r--')
plt.figure(figsize=(5,3), dpi=300)

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)
```

### BAR CHART


```python
labels=['A','B','C']
values=[1,4,2]
bars= plt.bar(labels,values , width=0.8 , color='yellow')

plt.figure(figsize=(2,1))


# design inside bars;
bars[0].set_hatch('/')
bars[1].set_hatch('+')
bars[2].set_hatch('*')



```

### BAR CHART WITH STRING VALUES

```python
company=['GOOGLE','AMAZON','OLA','FB']
revenue=[90,136,89,27]
```

```python
plt.xlabel('Companies')
plt.ylabel('US Dollar in Billion')
plt.title('US Technology Stocks')

plt.bar(company,revenue , label="Revenue" , width=0.4, color='#aabbcc')
plt.show()
```

### Multiple Bars 

```python
company=['GOOGLE','AMAZON','OLA','FB']
revenue=[90,136,89,27]
profit=[40,90,20,12]
```

```python
#here we have to assign xposition and then make adjustments like +_0.2 to make two diff charts 

xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')

plt.bar(xpos-0.2,revenue, width=0.4, color='#a1e87b' , label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4, color='#2e8003' , label="Profit")
plt.legend()
plt.show()
```

### Horizontal bar chart using barh function

```python
company=['GOOGLE','AMAZON','OLA','FB']
revenue=[90,136,89,27]
```

```python

plt.xlabel('US Dollar in Billion')
plt.title('US Technology Stocks')

plt.barh(company,revenue ,height=0.5, color='#8a3e7c',label="Revenue")
plt.show()
```
