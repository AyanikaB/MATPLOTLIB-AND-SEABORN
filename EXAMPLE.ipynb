```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

### LINE CHART


```python
plt.plot(gas.Year , gas.Canada , 'b.-' , label='Canada')
plt.plot(gas.Year , gas.USA , 'y.-' ,label='USA')
plt.plot(gas.Year , gas.France , 'g.-',label='France')
plt.plot(gas.Year , gas['South Korea'] , 'r.-', label='South Korea')          #south korea two words so ['']

plt.legend()
plt.title('Gas Prices over the Years' ,fontdict={'fontweight':'bold', 'fontsize': 16})
plt.xlabel('Year')
plt.ylabel('US Dollars')
plt.xticks(gas.Year[::3])    #every 3 years apart
plt.show()
plt.figure(figsize=(8,5))
plt.savefig('Gas_price_figure.png', dpi=300)
```

```python
# Another Way to plot many values!
countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
 if country in countries_to_look_at:
     plt.plot(gas.Year, gas[country], marker='.')
```

```python
for country in gas:
    if country !='Year':
        plt.plot(gas.Year, gas[country], marker='.')
```

### FIFA DATA

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

```python
bins=[30,40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins , color='#40c98a')
plt.xticks(bins)
plt.xlabel('Skill level')
plt.ylabel('No of Players')
plt.title('Distribution of Player Skills in FIFA')
plt.show()
```

### PIE CHART

```python
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
right
```

```python
left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels=['Left','Right']
color= ['#40c98a' , '#abcdef']

plt.pie([left,right] , labels=labels , colors=color ,autopct='%.2f %%')
plt.title('Foot Preference of FIFA Players')
plt.show()
```

### PIE CHART #2

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight    #to strip off the word lbs and make it an int
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')       #colour pattern

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)     #explodes the chart intodiff pieces

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

plt.show()
```
