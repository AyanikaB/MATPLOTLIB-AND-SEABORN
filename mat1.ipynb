```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(x,y,color='yellow',linewidth=5,linestyle='dotted')
```
