```python
import seaborn as sns
```

```python
sns.get_dataset_names()                  
#we can load datasets in seaborn itself.These are the names of the datasets that exists here. 
```

```python
tips= sns.load_dataset('tips')    #for the sake of learning we will load a few from seaborn itself.
iris= sns.load_dataset('iris')
titanic= sns.load_dataset('titanic')
planets= sns.load_dataset('planets')
```

```python
tips
```

```python
sns.scatterplot(x='tip' , y='total_bill' , data=tips , hue='day' , size='size')

#hue -
#Grouping variable that will produce points with different colors. Can be either categorical or numeric.

#size - 
#Grouping variable that will produce points with different sizes. Can be either categorical or numeric.
```

```python
sns.histplot(tips['tip'] , kde=True , color= '#aabbcc', bins=15) 

#histogram. kde indicates the line; remove kde and that line will be gone.
```

```python
sns.displot(tips['tip'] , kde=True , color= 'green', bins=10)

#distribution plot same as above
```

```python
sns.barplot(x='sex', y='tip' , data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips, hue='sex' , palette='YlGnBu')
```

```python
sns.stripplot(x='day', y ='tip',data=tips, dodge=True, hue='sex', palette='YlGnBu')
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips , kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips , kind='kde', fill=True)
```

```python
sns.jointplot(x='tip',y='total_bill', data=tips , kind='hex')
```

### pairplot - 
Plot pairwise relationships in a dataset.
By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column.

```python
sns.pairplot(titanic.select_dtypes(['number']), hue='pclass')
```
